```python
# 导库
import networkx as nx

# 输入边和权
s = [f"$v_{i}$" for i in [1, 1, 2, 2, 3]]
t = [f"$v_{i}$" for i in [3, 4, 3, 4, 4]]
w = [10, 60, 5, 20, 1]
edges = list(zip(s, t, w))

# 创建空图并添加顶点和边权
G = nx.Graph()
G.add_weighted_edges_from(edges)

# 计算顶点位置
pos = nx.spring_layout(G)

# 绘制无权图
nx.draw(G, pos, with_labels=True,font_size=14)

# 追加绘制权
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels,font_color="red")

# 导库
import networkx as nx

# 输入边和权
s = [f"$v_{i}$" for i in [1, 1, 2, 2, 3]]
t = [f"$v_{i}$" for i in [3, 4, 3, 4, 4]]
w = [10, 60, 5, 20, 1]
edges = list(zip(s, t, w))

# 创建空图并添加顶点和边权
G = nx.Graph()
G.add_weighted_edges_from(edges)

# 计算顶点位置
pos = nx.spring_layout(G)

# 绘制无权图
nx.draw(G, pos, with_labels=True,font_size=14)

# 追加绘制权
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels,font_color="red")
```

```python
# 导库
import networkx as nx
import numpy as np

# 输入邻接矩阵
A = np.array([[0, 0, 10, 60], [0, 0, 5, 20], [10, 5, 0, 1], [60, 20, 1, 0]])
G = nx.from_numpy_array(A,parallel_edges=False)

# 计算顶点位置
pos = nx.spring_layout(G)

# 绘制无权图
nx.draw(G,pos,with_labels=True)

# 追加绘制权
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels,font_color="red")
```
